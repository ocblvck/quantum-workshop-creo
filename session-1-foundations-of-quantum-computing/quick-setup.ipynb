## Step 4: Save Your Credentials

Now we'll save your API token to connect to IBM Quantum computers.

### ‚ö†Ô∏è IMPORTANT Instructions:

1. **Replace** `'YOUR_API_TOKEN_HERE'` with your actual token (keep the quotes!)
2. **Run this cell ONLY ONCE** - your credentials will be saved permanently
3. **After running successfully**, you can delete your token from this cell for security

### What This Does:

- Saves your credentials securely on your computer
- You won't need to enter it again for future quantum programs
- Enables access to real quantum hardware

**Ready?** Replace the token and run the cell below:

````python
# Replace 'YOUR_API_TOKEN_HERE' with your actual token from quantum.cloud.ibm.com
YOUR_API_TOKEN = "YOUR TOKEN"  # ‚Üê Paste your token between the quotes

# Run this cell to save your credentials
try:
    # Save account credentials (one-time setup)
    QiskitRuntimeService.save_account(
        channel="ibm_quantum_platform",  # Updated channel name
        token=YOUR_API_TOKEN,
        overwrite=True
    )
    print("="*60)
    print("‚úÖ SUCCESS! Account saved successfully!")
    print("="*60)
    print("\nüéâ You can now access IBM Quantum computers!")
    print("\nüí° TIP: You can now delete your token from this cell")
    print("   Your credentials are saved and you won't need it again.")
    print("="*60)
    
except Exception as e:
    if "YOUR_API_TOKEN" in YOUR_API_TOKEN:
        print("="*60)
        print("‚ùå ERROR: Token not replaced!")
        print("="*60)
        print("\nüìù You need to replace 'YOUR_API_TOKEN_HERE' with your actual token!")
        print("\nüîó Get your token from: https://quantum.cloud.ibm.com")
        print("   ‚Üí Log in ‚Üí Click profile ‚Üí Copy API token")
        print("="*60)
    else:
        print("="*60)
        print(f"‚ùå Error saving account: {e}")
        print("="*60)
        print("\nüí° Possible issues:")
        print("   1. Token might be incorrect - check for typos")
        print("   2. Internet connection issue")
        print("   3. IBM Quantum service might be down")
        print("\nüîó Get a new token from: https://quantum.cloud.ibm.com")
        print("="*60)
````

## Step 5: Test Your Connection (Optional)

Let's verify that everything is working by listing available quantum backends!

This will show you:
- Available quantum computers
- Number of qubits each has
- Current queue status

Run the cell below to see what quantum hardware you can access:

````python
# Test connection and list available quantum backends
try:
    # Load your saved credentials (must match the channel used in save_account)
    service = QiskitRuntimeService(channel="ibm_quantum_platform")
    
    print("="*60)
    print("‚úÖ Successfully connected to IBM Quantum!")
    print("="*60)
    
    # Try to get backends
    print("\nüîç Checking for available backends...")
    
    try:
        # Get list of all backends without any filters
        backends = service.backends()
        
        if len(backends) == 0:
            print("\nüìã No backends currently available.")
            print("\nüí° This is normal for new accounts. You can:")
            print("   ‚Ä¢ Use local simulators (see next cell)")
            print("   ‚Ä¢ Wait for account activation to complete")
            print("   ‚Ä¢ Check IBM Quantum Platform for access status")
        else:
            print(f"\nüåê Found {len(backends)} backend(s):")
            print("="*60)
            
            for i, backend in enumerate(backends, 1):
                try:
                    config = backend.configuration()
                    status = backend.status()
                    
                    print(f"\n{i}. {backend.name}")
                    print(f"   ‚Ä¢ Qubits: {config.n_qubits}")
                    print(f"   ‚Ä¢ Operational: {'üü¢ Yes' if status.operational else 'üî¥ No'}")
                    if hasattr(status, 'pending_jobs'):
                        print(f"   ‚Ä¢ Pending jobs: {status.pending_jobs}")
                except Exception as backend_error:
                    print(f"\n{i}. {backend.name} (limited info)")
            
            print("\n" + "="*60)
            print("üéâ You're ready to run quantum circuits!")
            print("="*60)
            
    except Exception as backend_error:
        print(f"\n‚ö†Ô∏è Could not retrieve backends: {backend_error}")
        print("\nüí° This is common with new accounts. No worries!")
        print("   You can still proceed with local simulators.")
    
    print("\nüí° REMEMBER: Simulators work without backend access!")
    print("   from qiskit_aer import AerSimulator")
    print("   simulator = AerSimulator()")
    print("="*60)
    
except Exception as e:
    print("="*60)
    print("‚ùå Connection Error")
    print("="*60)
    print(f"\nError: {e}")
    print("\nüí° Troubleshooting steps:")
    print("   1. Verify your account at: https://quantum.cloud.ibm.com")
    print("   2. Make sure you've accepted terms of service")
    print("   3. Check if your account is fully activated")
    print("   4. Try regenerating your API token")
    print("\n‚úÖ Good news: You can still use local simulators!")
    print("   The setup was successful for simulator access.")
    print("="*60)
````

## Alternative: Test with a Simple Circuit

If you're having issues with backend access, you can still use simulators! Run this cell to verify your Qiskit installation works:

````python
# Test with a simple quantum circuit using local simulator
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

# Create a simple Bell state circuit
qc = QuantumCircuit(2, 2)
qc.h(0)
qc.cx(0, 1)
qc.measure([0, 1], [0, 1])

print("‚úÖ Simple Quantum Circuit Created!")
print("\n" + str(qc.draw(output='text')))

# Run on local simulator
simulator = AerSimulator()
job = simulator.run(qc, shots=1000)
result = job.result()
counts = result.get_counts()

print("\nüìä Simulation Results:")
print(counts)
print("\nüéâ Your Qiskit installation is working perfectly!")
print("   You can now proceed with the workshop notebooks.")
````
